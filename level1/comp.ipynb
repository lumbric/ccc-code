```python
file = "level1_2_large.in"

with open(file) as f:
    c = f.read()

e = c.split("\n")
```

```python
s = []
for i in range(1, len(e) - 1):
    s.append(sum([int(ee) for ee in e[i].split(' ')]))
```

```python
with open("test", "w") as f:
    c = f.write("\n".join([str(e) for e in s]))
    f.write("\n")
```
