```python
from pathlib import Path
import numpy as np

LEVEL_NR = 4

file_in = Path(f"level{LEVEL_NR}_0_example.in")
file_out = file_in.with_suffix(".out-test")

file_out
```

```python
with open(file_in) as f:
    content = f.read()

lines = content.split("\n")

scenarios = []
for line in lines[1:-1]:
    space_station,time_limit = line.split(' ')
    scenarios.append([tuple(int(coord) for coord in space_station.split(',')), int(time_limit)])

scenarios
```

```python
def total_time(paces):
    return np.abs(paces).sum() + (paces == 0).sum()


def pos_after(paces):
    return np.sign(paces).sum()
```

```python
def navigate_to_space_station(space_station, time_limit):
    target_pos = abs(space_station)

    target_posp = int(np.abs(target_pos))
    # print(target_pos)

    l = [5, 4, 3, 2, 1]

    acc_pc = int(np.floor(target_posp/2))
    acc_pc

    min(len(l), acc_pc)

    acc_s = l[0:int(min(len(l), acc_pc))]
    dcc_s = acc_s[::-1]

    cs = max(min(acc_s) - 1, 1)

    return [0] + acc_s + [cs] * (target_posp - 2*len(acc_s)) + dcc_s + [0]

```

```python
def navigate_to_space_station_2d(space_station, time_limit):
    x_target, y_target = space_station
    paces_x = navigate_to_space_station(x_target, time_limit)
    paces_y = navigate_to_space_station(y_target, time_limit)

    return paces_x, paces_y
```

```python
with open(file_out, "w") as f:
    for scenario in scenarios:
        result_x,result_y = navigate_to_space_station_2d(*scenario)
        content = f.write(" ".join([str(pace) for pace in result_x]) + "\n")
        content = f.write(" ".join([str(pace) for pace in result_y]) + "\n")
        
        f.write("\n")
```
