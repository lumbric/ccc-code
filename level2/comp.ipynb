```python
from pathlib import Path
import numpy as np

file_in = Path("level2_0_example.in")
file_out = file_in.with_suffix(".out-test")

file_out
```

```python
with open(file_in) as f:
    c = f.read()

e = c.split("\n")
e
```

```python
s = []
for i in range(1, len(e) - 1):
    s.append(np.array([int(ee) for ee in e[i].split(' ')]))
```

```python
def compute(a):
    return [np.sign(a).sum(), np.abs(a).sum() + (a == 0).sum()]

[compute(e) for e in s]
```

```python
with open(file_out, "w") as f:
    for e in s:
        p, t = compute(e)
        c = f.write(f"{p} {t}\n")
    f.write("\n")
```
